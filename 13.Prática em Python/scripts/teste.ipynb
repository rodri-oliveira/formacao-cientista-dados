```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as srn
```

```python
base = pd.read_csv("../dados/dados.csv", sep=";")
base.shape
```

```python
base.head()
```

```python
srn.histplot(base.PIB, kde=True, bins=10).set(title='PIB')
```

```python
srn.histplot(base.VALOREMPENHO, kde=True, bins=10).set(title='Empenho')
```

```python
agrupado = base.sort_values('PIB').head(10)
agrupado = agrupado.iloc[:,1:3]
agrupado
agrupado.plot.bar(x='MUNICIPIO', y='PIB', color='gray')
```

```python
agrupado = base.sort_values('PIB', ascending=False).head(10)
agrupado = agrupado.iloc[:,1:3]
agrupado
agrupado.plot.bar(x='MUNICIPIO', y='PIB', color='gray')
```

```python
# Criar o grfico de barras
plt.figure(figsize=(10, 6))  # Define o tamanho da figura
plt.bar(agrupado['MUNICIPIO'], agrupado['PIB'], color='skyblue')  # Personalize as cores se desejar
plt.xlabel('Municpio')
plt.ylabel('PIB')
plt.title('Top 10 Municpios com os Maiores PIBs')

# Rotacionar os rtulos do eixo x para melhor legibilidade
plt.xticks(rotation=45, ha='right')

# Exibir os valores acima das barras
for i, v in enumerate(agrupado['PIB']):
    plt.text(i, v, f'{v:.2f}', ha='center', va='bottom', fontsize=10)

plt.tight_layout()  # Melhora a disposio do grfico
plt.show()
```
